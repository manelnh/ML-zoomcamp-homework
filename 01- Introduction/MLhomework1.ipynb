```python
asian_cars = df[df['origin'] == 'Asia']
asian_cars
```

```python
selected_columns = asian_cars[['vehicle_weight', 'model_year']]
selected_columns
```

```python
data = selected_columns.head(7)
data
```

```python
X = np.array(data)
X
```

```python
XTX = X.T
XTX
```

```python
XTX_invert = np.linalg.inv(XTX @ X)
XTX_invert
```

```python
y = np.array([1100, 1300, 800, 900, 1000, 1100, 1200])
```

```python
w = XTX_invert @ XTX @ y
w
```

```python
result = w.sum()
result
```
